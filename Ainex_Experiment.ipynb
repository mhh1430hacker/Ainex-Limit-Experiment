```python
# ==============================================================================
# مشروع قانون آينكس: السيمفونية الأخيرة (20 Generations)
# ==============================================================================

!pip install transformers datasets sentence-transformers scipy scikit-learn accelerate matplotlib seaborn > /dev/null

import torch
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from datasets import load_dataset
from sentence_transformers import SentenceTransformer
from sklearn.decomposition import PCA
from scipy.spatial import ConvexHull, distance_matrix
import numpy as np
import random
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import textwrap

warnings.filterwarnings("ignore")

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Working on: {device.upper()}")

# تثبيت العشوائية لنتائج قابلة للتكرار
seed_val = 42
random.seed(seed_val)
np.random.seed(seed_val)
torch.manual_seed(seed_val)
if device == 'cuda':
    torch.cuda.manual_seed_all(seed_val)

# ==============================================================================
# إعداد النماذج والبيانات
# ==============================================================================

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token
model = GPT2LMHeadModel.from_pretrained("gpt2").to(device)
embedder = SentenceTransformer('all-MiniLM-L6-v2')

# تحميل البيانات البشرية (Baseline)
print("... جاري تحميل بيانات البشر ...")
dataset = load_dataset("wikitext", "wikitext-2-raw-v1", split="train")
# نأخذ عينة أصغر قليلاً لتسريع الـ 20 جيل
real_wiki_texts = [x['text'] for x in dataset if len(x['text']) > 100][:250]

# النص الشاهد الثابت (سنرى كيف يتغير توليده)
witness_prompt = "The fundamental laws of physics dictate that"
witness_history = []

class TextListDataset(Dataset):
    def __init__(self, texts, tokenizer, max_length=128):
        self.inputs = []
        for txt in texts:
            if len(txt) > 20:
                enc = tokenizer(txt, truncation=True, max_length=max_length, padding="max_length", return_tensors="pt")
                self.inputs.append(enc)
    def __len__(self): return len(self.inputs)
    def __getitem__(self, i): return {k: v.squeeze(0) for k, v in self.inputs[i].items()}

# ==============================================================================
# دوال المحرك والقياس
# ==============================================================================

def train_model(model, texts, epochs=2): # تقليل الـ epochs قليلاً لتسريع العملية
    model.train()
    train_data = TextListDataset(texts, tokenizer)
    # batch size أكبر قليلاً للتسريع
    loader = DataLoader(train_data, batch_size=16, shuffle=True)
    optim = AdamW(model.parameters(), lr=5e-5)

    for epoch in range(epochs):
        loop = tqdm(loader, leave=False, desc=f"Training Epoch {epoch+1}")
        for batch in loop:
            optim.zero_grad()
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            outputs = model(input_ids, attention_mask=attention_mask, labels=input_ids)
            loss = outputs.loss
            loss.backward()
            optim.step()

def generate_data(model, num_samples=250, prompt_override=None):
    model.eval()
    gen_texts = []
    if prompt_override:
        prompts = [prompt_override]
        num_samples = 1
    else:
        prompts = ["The history", "Science is", "War began", "The theory", "He was born",
                   "In 1990", "The system", "Water is", "Computers are", "The city"] * (num_samples // 10 + 5)

    batch_size = 25 if not prompt_override else 1

    loop_iter = tqdm(range(0, num_samples, batch_size), leave=False) if not prompt_override else range(1)

    for i in loop_iter:
        batch_prompts = prompts[i:i+batch_size] if not prompt_override else prompts
        if not batch_prompts: break
        inputs = tokenizer(batch_prompts, return_tensors="pt", padding=True, truncation=True).to(device)
        with torch.no_grad():
            outputs = model.generate(
                **inputs, max_length=100, do_sample=True, top_k=50, temperature=0.8, repetition_penalty=1.05
            )
        for out in outputs:
            gen_texts.append(tokenizer.decode(out, skip_special_tokens=True))

    if prompt_override: return gen_texts[0]
    return gen_texts[:num_samples]

# --- دالة حساب المقاييس الخمسة ---
def calculate_ainex_metrics(human_texts, gen_texts):
    human_emb = embedder.encode(human_texts)
    gen_emb = embedder.encode(gen_texts)

    pca = PCA(n_components=3)
    pca.fit(human_emb)
    h_coords = pca.transform(human_emb)
    g_coords = pca.transform(gen_emb)

    # 1. الإبداع (Volume)
    try: vol = ConvexHull(g_coords).volume
    except: vol = 0.0

    # 2. الجنون/الهلوسة (Drift from Human Center)
    human_centroid = np.mean(h_coords, axis=0)
    gen_centroid = np.mean(g_coords, axis=0)
    drift = np.linalg.norm(human_centroid - gen_centroid)

    # 3. التنوع (Dispersion - متوسط البعد عن مركز الجيل نفسه)
    dispersion = np.mean(np.linalg.norm(g_coords - gen_centroid, axis=1))

    # 4. الضغط (Density - متوسط المسافات بين النقاط)
    # كلما قل الرقم، زاد "الضغط" والتكرار
    dist_mat = distance_matrix(g_coords, g_coords)
    # نأخذ متوسط المسافات (نستثني الأصفار القطرية)
    density = np.mean(dist_mat[dist_mat > 0]) if dist_mat.sum() > 0 else 0

    # 5. الهلوسة الشاذة (Max Outlier Distance - أبعد نقطة عن المركز)
    hallucination_max = np.max(np.linalg.norm(g_coords - gen_centroid, axis=1))

    return {
        "Creativity (Volume)": vol,
        "Madness (Drift)": drift,
        "Diversity (Dispersion)": dispersion,
        "Compression (Avg Dist)": density,
        "Hallucination (Max Outlier)": hallucination_max,
        "Coords": g_coords # نحفظ الإحداثيات للرسم
    }

# ==============================================================================
# التجربة الكبرى (20 Generations)
# ==============================================================================

metrics_history = {
    "Creativity (Volume)": [], "Madness (Drift)": [], "Diversity (Dispersion)": [],
    "Compression (Avg Dist)": [], "Hallucination (Max Outlier)": []
}
point_clouds_history = []

print("\n--- Measuring Human Baseline (Gen 0) ---")
baseline_metrics = calculate_ainex_metrics(real_wiki_texts, real_wiki_texts)
for k, v in baseline_metrics.items():
    if k != "Coords": metrics_history[k].append(v)
point_clouds_history.append(baseline_metrics["Coords"])

# توليد النص الشاهد الأول (البشري تقريباً)
witness_text = generate_data(model, prompt_override=witness_prompt)
witness_history.append(f"Gen 0 (Base): {witness_text}")
print(f"Witness Text (Gen 0): {witness_text[:100]}...")

current_data = real_wiki_texts
total_generations = 20 # الرحلة الطويلة

for gen in range(1, total_generations + 1):
    print(f"\n" + "="*40)
    print(f" STARTING GENERATION {gen} / {total_generations}")
    print("="*40)

    # 1. تدريب
    train_model(model, current_data, epochs=2)

    # 2. توليد بيانات الجيل
    new_data = generate_data(model, num_samples=250)

    # 3. توليد النص الشاهد
    witness_text = generate_data(model, prompt_override=witness_prompt)
    witness_history.append(f"Gen {gen:02d}: {witness_text}")

    # 4. حساب المقاييس
    gen_metrics = calculate_ainex_metrics(real_wiki_texts, new_data)
    for k, v in gen_metrics.items():
        if k != "Coords": metrics_history[k].append(v)
    point_clouds_history.append(gen_metrics["Coords"])

    print(f">> Gen {gen} Metrics: Vol={gen_metrics['Creativity (Volume)']:.3f}, Drift={gen_metrics['Madness (Drift)']:.3f}")

    current_data = new_data

# ==============================================================================
# العرض النهائي (The Grand Reveal)
# ==============================================================================

# 1. عرض النص الشاهد (كيف جن النموذج)
print("\n" + "#"*50)
print(" THE WITNESS TEXT EVOLUTION (Human -> Madness) ")
print("#"*50)
top_gens = [0, 1, 5, 10, 15, 20] # نختار محطات رئيسية
for i in top_gens:
    if i < len(witness_history):
        print(f"\n--- {witness_history[i].split(':')[0]} ---")
        print(textwrap.fill(witness_history[i].split(':', 1)[1].strip(), width=80))

# 2. رسم لوحة القيادة (5 رسوم بيانية)
plt.style.use('dark_background')
fig, axes = plt.subplots(5, 1, figsize=(10, 18), sharex=True)
gens_axis = range(total_generations + 1)
colors = ['#00ff00', '#ff0000', '#ffff00', '#00ffff', '#ff00ff'] # ألوان مميزة

for i, (metric_name, values) in enumerate(metrics_history.items()):
    ax = axes[i]
    ax.plot(gens_axis, values, marker='o', linestyle='-', color=colors[i], linewidth=2)
    ax.fill_between(gens_axis, values, color=colors[i], alpha=0.2)
    ax.set_ylabel(metric_name, fontsize=10, color=colors[i])
    ax.grid(True, alpha=0.3)
    if i == 0: ax.set_title('The Ainex Limit: 20-Generation Collapse Dashboard', fontsize=14, fontweight='bold')
    if i == 4: ax.set_xlabel('Generation', fontsize=12)

plt.tight_layout()
plt.savefig('ainex_20gen_dashboard.png', dpi=300)
print("\n>> تم حفظ لوحة القيادة: 'ainex_20gen_dashboard.png'")

# 3. رسم شبكة النقاط ثلاثية الأبعاد (البداية والنهاية)
fig_3d = plt.figure(figsize=(14, 6))

# Gen 0 (Human)
ax1 = fig_3d.add_subplot(1, 2, 1, projection='3d')
p0 = point_clouds_history[0]
ax1.scatter(p0[:,0], p0[:,1], p0[:,2], c='#00ff00', s=20, alpha=0.6, label='Human (Gen 0)')
ax1.set_title("Gen 0: Human Knowledge (Structured)", color='white')
ax1.set_axis_off()

# Gen 20 (Madness)
ax2 = fig_3d.add_subplot(1, 2, 2, projection='3d')
p20 = point_clouds_history[-1]
# تلوين حسب البعد عن المركز لإظهار الهلوسة
dist_from_center = np.linalg.norm(p20 - np.mean(p20, axis=0), axis=1)
ax2.scatter(p20[:,0], p20[:,1], p20[:,2], c=dist_from_center, cmap='inferno', s=30, alpha=0.8, label='AI (Gen 20)')
ax2.set_title("Gen 20: AI Madness (Chaotic & Drifted)", color='white')
ax2.set_axis_off()

plt.savefig('ainex_20gen_pointcloud.png', dpi=300)
print(">> تم حفظ شبكة النقاط: 'ainex_20gen_pointcloud.png'")

plt.show()
```
